```python
def print_difference(page_names_api, page_names_report):
    print('{} pages are both in the report and found by the API'.format(len(page_names_api & page_names_report)))

    not_in_report = page_names_api - page_names_report
    print()
    print('{} pages are found by the API but do not appear in the report:'.format(len(not_in_report)))
    print(not_in_report)

    not_in_api = page_names_report - page_names_api
    print()
    print('{} pages are found in the report but not by the API:'.format(len(not_in_api)))
    print(not_in_api)
```

```python
print_difference(page_names_api=page_names_api, page_names_report=page_names_report)
```
